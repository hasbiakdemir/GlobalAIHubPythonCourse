```python
#key'lerin username, value'ların password olduğu bir dictionary tanımlanır.
userDB = {"deniz": "A12345", "seckin": "B98765", "merve": "C75395", "ibrahim": 175395, "mustafa": 798521}
print(userDB, type(userDB))
```

```python
userDB.items()
```

```python
username = input("Username: ").strip()

print("Press q to Quit")
password = input("Password: ").strip()

#Username and Password Identity Check
while True:
    if password == 'q':
        break
    elif (username not in userDB.keys()):
        print("1-1-Check Username or Password")
        username = input("Username: ").strip()
        password = input("Password: ").strip()
    else:
        break

#User Login Application
for u, p in userDB.items():
    while (u == username):
        if (p == password):
            print("1-1-You Entered Successfully!")
            break
        elif (p != password): #Counter before Reset
            x = 0
            while (x < 2):
                print("1-2-Check Username or Password")
                username = input("Username: ").strip()
                password = input("Password: ").strip()
                if (username == 'q' or password == 'q'):
                    break
                elif (p == password):
                    print("1-2-You Entered Successfully")
                    break
                x += 1
                #Username and Password Identity Check
                while True:
                    if (username == 'q' or password == 'q'):
                        break
                    elif (username not in userDB.keys()):
                        print("1-3-Check Username or Password")
                        username = input("Username: ").strip()
                        password = input("Password: ").strip()
                    else:
                        break
            else:
                if (p == password):
                    print("1-3-You Entered Successfully!")
                    break
                else:
                    print("Reset Your Password!")
                    break
        break
```

```python
n = int(input("Tek Basamakli Bir Sayi Giriniz: "))

cnt = 1 #start_count
max_cnt = 3 #maximum try limit

while (cnt <= max_cnt): #counter
    if ((max_cnt - cnt == 0)): #last try
        if (n >= 0 and n < 10 and n % 2 == 0): #tek haneli ift say ise
            lst = list(range(0, n+2, 2))
            print("Sonuc:", lst)
            break
        elif (n >= 0 and n < 10 and n % 2 != 0): #tek haneli tek say ise
            lst = list(range(0, n, 2))
            print("Sonuc:", lst)
            break
        else:
            print("Deneme Sayisi Asildi!")
            break
    elif (max_cnt - cnt == 1): #penultimate try
        if (n >= 0 and n < 10 and n % 2 == 0): #tek haneli ift say ise
            lst = list(range(0, n+2, 2))
            print("Sonuc:", lst)
            break
        elif (n >= 0 and n < 10 and n % 2 != 0): #tek haneli tek say ise
            lst = list(range(0, n, 2))
            print("Sonuc:", lst)
            break
        elif (n < 0): #girilen deer negatif ve son deneme hakk ise
            print("2-Pozitif Bir Sayi Girmelisiniz!")
            print("Son Deneme Hakkiniz!")
            cnt += 1
            n = int(input("Lutfen Tekrar Deneyiniz: "))
        else: #girilen deer tek haneli deil ve son deneme hakk ise
            print("2-Tek Haneli Bir Sayi Girmediniz")
            print("Son Deneme Hakkiniz!")
            cnt += 1
            n = int(input("Lutfen Tekrar Deneyiniz: "))
    else:
        if (n >= 0 and n < 10 and n % 2 == 0): #tek haneli ift say ise
            lst = list(range(0, n+2, 2))
            print("Sonuc:", lst)
            break
        elif (n >= 0 and n < 10 and n % 2 != 0): #tek haneli tek say ise
            lst = list(range(0, n, 2))
            print("Sonuc:", lst)
            break
        elif (n < 0): #girilen deer negatif ancak son deneme hakk deil ise
            print("1-Pozitif Bir Sayi Girmelisiniz!")
            print(f"Kalan Deneme Says: {max_cnt - cnt}")
            cnt += 1
            n = int(input("Lutfen Tekrar Deneyiniz: "))
        else: #girilen deer tek haneli deil ve son deneme hakk deil ise
            print("1-Tek Haneli Bir Sayi Girmediniz")
            print(f"Kalan Deneme Says: {max_cnt - cnt}")
            cnt += 1
            n = int(input("Lutfen Tekrar Deneyiniz: "))
```

```python
####################### ÖĞRENCİ NOT GİRİŞİ ################
def student():
    x = input("ID: ").strip()

#Username and Password Identity Check
    while True:
        if (x not in studentID.keys()):
            print("Check Your Student ID")
            x = input("ID: ").strip()
        elif (x in lst_temp): ########lst_temp olarak aşağıda kullanılması önemli ######
            print("Scores for this student are already entered!")
            x = input("ID: ").strip()
        else:
            return x, studentID[x]

################# PASSING GRADE CALCULATION #######################
def passGrade(*args):
    return round(args[0] * (0.3) + args[1] * (0.3) + args[2] * (0.4))

################ NOT GİRİŞLERİ ####################################
def grades():
    while True:
        try:
            value = int(input())
        except ValueError:
            print("Lutfen 0 ile 100 arasinda bir deger girin")
            continue

        if value < 0:
            print("Girilen deger negatif olamaz.")
            continue
        elif value > 100:
            print("Girilen deger 100'den buyuk olamaz.")
            continue
        else:
            break
    return value
```

```python
studentID = {"123456": "Deniz Yılmaz", "987654": "Ahmet Demir", "741258": "Ali Veli", "369852": "Deniz Demir",
             "951753": "Mehmet Yılmaz"}
```

```python
################## NOT GİRİŞİ (Midterm - Project - Final sırasıyla) ##########################
lst = [None] * len(studentID) 
lst_temp = [] #ismi değiştirilmemeli
for i in range(len(studentID)):
    lst[i] = [student()] + [grades(), grades(), grades()]
    i += 1
```

```python
################### PASSING GRADE HESAPLAMA ############################
passingGrade = [None] * len(studentID)
for i in range(len(studentID)):
    passingGrade[i] = [passGrade(lst[i][1], lst[i][2], lst[i][3])]
    i += 1
```

```python
######################## SON DURUM ##########################
lsst = [None] * len(studentID)
for i in range(len(studentID)):
    lsst[i] = lst[i]+passingGrade[i]
    i += 1
```

```python
################### DICTIONARY ÇEVİRİMİ ################################3
def convert(lst):
    res_dct = {lst[i][0][0]: [lst[i][0][1], [lst[i][1], lst[i][2], lst[i][3]], lst[i][4]] for i in range(0, len(lst), 1)}
    return res_dct
```

```python
studentDB = convert(lsst)
studentDB
```

```python
################ DICTIONARY - LIST ÇEVRİMİ ##################################
studentList = []
temp = []

for key, value in studentDB.items():
    temp = [key,value]
    studentList.append(temp)
```

```python
######################## EN DÜŞÜK VE EN YÜKSEK NOTLU ÖĞRENCİYİ BULMA #######################
u = []

for i in range(len(studentList)):
    u.append([i, studentList[i][1][2]])
    i += 1

u
```

```python
########index - final notu liste içerisinde liste olması sebebiyle iki ayrı liste olarak parçalandı.
p1 = []
p2 = []
for i in range(len(u)):
    p1.append(u[i][0])
    i += 1

for j in range(len(u)):
    p2.append(u[j][1])
    j += 1
    
p1, p2
```

```python
############# iki liste içerisindeki minimum ve maksimum değerler bulundu.
minimum_v, maksimum_v = min(p2), max(p2)
minimum_v, maksimum_v
```

```python
############# maksimum ve minimum değerlerin index'i bulundu ########################3
minimum_i, maksimum_i = p2.index(minimum_v), p2.index(maksimum_v)
minimum_i, maksimum_i
```

```python
if (minimum_i < maksimum_i):
    minimum_i, maksimum_i = maksimum_i, minimum_i
else:
    minimum_i, maksimum_i = maksimum_i, minimum_i
```
